```python
import nbformat
from nbconvert import PythonExporter
from nbconvert.preprocessors import ExecutePreprocessor
import os
```

```python
mice = ["466029_f3", "475165_f1", "475165_f3", "475165_f5", "475165_f7"]
days = [f"day{d}" for d in range(1, 16)]

path_list = []
for mouse in mice:
    for day in days:
        path = f"C:/Users/trapped/Documents/GitHub/corticalanalysis/Notebook/2_stage2/third_cohort/auto/auto_{mouse}_stage2_{day}.ipynb"
        path_list.append(path)
```

```python
# Path to save executed notebooks
output_directory = "executed"
os.makedirs(output_directory, exist_ok=True)

# Execution function
def run_notebook(notebook_path, output_path):
    try:
        with open(notebook_path, "r", encoding="utf-8") as f:
            nb = nbformat.read(f, as_version=4)

        # Configure notebook execution
        ep = ExecutePreprocessor(timeout=600, kernel_name="python3")
        ep.preprocess(nb, {"metadata": {"path": os.path.dirname(notebook_path) or "./"}})

        # Save executed notebook
        with open(output_path, "w", encoding="utf-8") as f:
            nbformat.write(nb, f)
        print(f"Executed successfully: {notebook_path}")

    except Exception as e:
        print(f"Error while executing {notebook_path}: {e}")

# Execute each trial notebook sequentially
for notebook in path_list:
    output_path = os.path.join(output_directory, f"executed_{os.path.basename(notebook)}")
    print(f"Running: {notebook}")
    run_notebook(notebook, output_path)
```
